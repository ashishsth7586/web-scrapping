```python
from bs4 import BeautifulSoup
import requests
import csv
import datetime;
```

```python

source = requests.get('https://www.ktm2day.com/vegetables-price-in-nepali-market/').text
soup = BeautifulSoup(source, 'lxml')
```

```python
csv_file = open('./source/ktm2day_retailprice.csv', 'a')
csv_writer = csv.writer(csv_file)
csv_writer.writerow([f"{datetime.datetime.now():%Y-%m-%d}"])
csv_writer.writerow(['Product Name', 'Unit', 'Minimum', 'Maximum', 'Average'])
```

```python
price_table = soup.find('table')
tablerows = price_table.find_all('tr')[1:]
tablerows
```

```python
for tablerow in tablerows:
    tablerowdata = tablerow.text
    print(tablerowdata)
    products = tablerowdata.split('\n')[1:]
    csv_writer.writerow(products)
print()
```
